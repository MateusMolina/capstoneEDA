```python
#create a loop to go thru my sources and combine the articles in the past month:
sources = ['techcrunch', 'wired','hacker-news',
           "abc-news","ars-technica","axios","bloomberg", "google-news",
           "medical-news-today", "new-scientist", "recode", "techradar",
           "the-economist", "the-huffington-post", "the-new-york-times",
           "the-next-web", "the-verge", "the-wall-street-journal",
            "time",
          ]
store_dict = { 
    "date":[],
    "title":[],
    "source":[],
    "url":[]
}
for source in sources:
    print("Now querying ", source)
    #Paginates through a given query with a given source, adds it to the store_dict
    for page in range(1, 11):
        print("page: ", page)
        url = (f'https://newsapi.org/v2/everything?'
           'q="Machine Learning" OR "Artificial Intelligence"&'
           f'sources= {source}&'
           'from=2018-9-03&' # a month ago
           'to=2018-10-02&' # yesterday
           'pageSize=100&'
           f'page={page}&'
           f'apiKey={api_key}')
        response = requests.get(url)
        js_string = json.dumps(response.json())
        js_dict= json.loads(js_string)
        if len(js_dict['articles'])==0:
            time.sleep(1)
            break
        else:
            for art in js_dict['articles']:
                store_dict['date'].append(art['publishedAt'])
                store_dict['title'].append(art['title'])
                store_dict['source'].append(art['source']['id'])
                store_dict['url'].append(art['url'])
        time.sleep(1)
```

```python
news_10_02_18 = pd.DataFrame(store_dict)
```
